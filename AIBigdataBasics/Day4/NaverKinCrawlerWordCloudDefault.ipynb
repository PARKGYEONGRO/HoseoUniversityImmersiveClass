```python
import subprocess
import os
import re
import time
from collections import Counter
from urllib.parse import parse_qs, urlencode, urljoin, urlparse, urlunparse

import matplotlib
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import pandas as pd
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud

# ==========================================
# [설정 항목] 환경 변수 및 상수 정의
# ==========================================
START_URL = "https://kin.naver.com/qna/list.naver?dirId=70111&queryTime=2026-07-08%2009%3A05%3A31&page=1"
SAVE_FILE = "naver_kin_qna_with_detail.csv"
FONT_PATH = "/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf"

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    ),
    "Referer": "https://kin.naver.com/",
    "Accept-Language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
}

STOPWORDS = set([
    "그리고", "그러나", "그래서", "하지만", "또한", "이것", "저것", "그것",
    "제가", "저는", "나는", "너무", "정말", "진짜", "혹시", "계속", "합니다", "있습니다"
])


# ==========================================
# 1. 환경 설정 함수 (폰트 및 캐시 초기화)
# ==========================================
def init_korean_font(font_path=FONT_PATH):
    """리눅스(Colab 등) 환경에서 나눔 폰트를 설치하고 Matplotlib에 등록합니다."""
    print("[시스템] 한글 폰트 설정 및 초기화를 시작합니다...")
    
    # 폰트 설치 및 캐시 디렉터리 제거 (셀 명령어 실행)
    subprocess.run(["apt-get", "install", "-y", "fonts-nanum"])
    subprocess.run("rm -rf ~/.cache/matplotlib", shell=True)
    
    if os.path.exists(font_path):
        fm.fontManager.addfont(font_path)
        font_name = fm.FontProperties(fname=font_path).get_name()
        plt.rc("font", family=font_name)
        matplotlib.rcParams["axes.unicode_minus"] = False
        print(f"[시스템] 한글 폰트 설정 완료: {font_name}")
    else:
        # 대안 시스템 폰트 탐색
        found_font = None
        for file in fm.findSystemFonts(fontpaths=None, fontext="ttf"):
            if "NanumGothic" in file or "NanumBarunGothic" in file:
                found_font = file
                break
        
        if found_font:
            font_name = fm.FontProperties(fname=found_font).get_name()
            plt.rc("font", family=font_name)
            plt.rcParams["axes.unicode_minus"] = False
            print(f"[시스템] 시스템 검색을 통해 한글 폰트를 설정했습니다: {font_name}")
        else:
            print("[경고] 한국어 폰트를 찾지 못했습니다. 그래프 글자가 깨질 수 있습니다.")


# ==========================================
# 2. 크롤링 관련 유틸리티 및 헬퍼 함수
# ==========================================
def clean_text(text):
    if not text:
        return ""
    return re.sub(r"\s+", " ", text).strip()


def build_page_url(start_url, page):
    parsed = urlparse(start_url)
    query_dict = parse_qs(parsed.query)
    query_dict["page"] = [str(page)]
    new_query = urlencode(query_dict, doseq=True)
    return urlunparse((parsed.scheme, parsed.netloc, parsed.path, parsed.params, new_query, parsed.fragment))


def extract_doc_id(url):
    parsed = urlparse(url)
    query_dict = parse_qs(parsed.query)
    return query_dict.get("docId", [""])[0]


def get_text_by_selectors(soup, selectors, find_all=False):
    """지정된 셀렉터 목록 중 매칭되는 텍스트를 추출합니다."""
    texts = []
    for selector in selectors:
        if find_all:
            tags = soup.select(selector)
            for tag in tags:
                text = clean_text(tag.get_text(" ", strip=True))
                if text and text not in texts:
                    texts.append(text)
            if texts:
                break
        else:
            tag = soup.select_one(selector)
            if tag:
                text = clean_text(tag.get_text(" ", strip=True))
                if text:
                    return text
    return texts if find_all else ""


# ==========================================
# 3. 데이터 수집 핵심 비즈니스 로직
# ==========================================
def fetch_detail_text(session, detail_url, delay=0.8):
    """질문 상세 페이지 내 본문과 답변 텍스트를 수집합니다."""
    try:
        response = session.get(detail_url, timeout=10)
        response.raise_for_status()
    except requests.RequestException as e:
        print(f"    [상세 페이지 오류] {detail_url} -> {e}")
        return {"question_text": "", "answers_text": "", "detail_text": ""}

    soup = BeautifulSoup(response.text, "html.parser")

    question_selectors = [
        "div.c-heading__content", "div.c-heading-content", 
        "div.questionDetail", "div.endContentsText", 
        "div.se-main-container", "div._endContentsText"
    ]
    answer_selectors = [
        "div.answer-content__item div.se-main-container", 
        "div.answer-content__item", "div._endContentsText", 
        "div.endContentsText", "div.c-heading-answer__content-user"
    ]

    question_text = get_text_by_selectors(soup, question_selectors, find_all=False)
    answer_text_list = get_text_by_selectors(soup, answer_selectors, find_all=True)

    # 질문 본문이 답변 목록에 중복 수집되는 현상 방지
    answer_text_list = [text for text in answer_text_list if text != question_text]
    answers_text = "\n\n---답변 구분---\n\n".join(answer_text_list)
    detail_text = clean_text(question_text + " " + answers_text)

    time.sleep(delay)
    return {
        "question_text": question_text,
        "answers_text": answers_text,
        "detail_text": detail_text,
    }


def parse_list_page(html, page):
    """목록 페이지에서 개별 질문 링크와 메타데이터를 파싱합니다."""
    soup = BeautifulSoup(html, "html.parser")
    results = []
    seen = set()

    selectors = [
        "td.title a[href*='detail']", ".title a[href*='detail']",
        "dt a[href*='detail']", "a[href*='/qna/detail.naver']", "a[href*='/qna/detail.nhn']"
    ]

    links = []
    for selector in selectors:
        links = soup.select(selector)
        if links:
            break

    for a in links:
        title = clean_text(a.get_text(" ", strip=True))
        href = a.get("href", "")

        if not title or not href:
            continue

        full_url = urljoin("https://kin.naver.com", href)
        doc_id = extract_doc_id(full_url)

        if not doc_id or doc_id in seen:
            continue

        seen.add(doc_id)

        parent = a.find_parent(["tr", "li", "dl", "div"])
        parent_text = clean_text(parent.get_text(" ", strip=True)) if parent else ""

        date_match = re.search(r"\d{4}\.\d{1,2}\.\d{1,2}\.?", parent_text)
        date = date_match.group(0) if date_match else ""

        answer_match = re.search(r"답변\s*(\d+)", parent_text)
        answer_count = answer_match.group(1) if answer_match else ""

        results.append({
            "page": page,
            "doc_id": doc_id,
            "title": title,
            "url": full_url,
            "date": date,
            "answer_count": answer_count,
            "list_raw_text": parent_text,
        })

    return results


def crawl_kin_pipeline(start_url, total_pages=10, list_delay=1.2, detail_delay=0.8):
    """전체 크롤링 프로세스를 제어하는 메인 파이프라인 함수입니다."""
    all_data = []
    session = requests.Session()
    session.headers.update(HEADERS)

    for page in range(1, total_pages + 1):
        url = build_page_url(start_url, page)
        print(f"[목록 수집] page={page}/{total_pages} | url={url}")

        try:
            response = session.get(url, timeout=10)
            response.raise_for_status()
        except requests.RequestException as e:
            print(f"[오류] 목록 {page}페이지 접근 실패: {e}")
            continue

        list_data = parse_list_page(response.text, page)
        print(f" -> {len(list_data)}개의 항목 발견")

        for idx, item in enumerate(list_data, start=1):
            print(f"    [상세 수집] ({idx}/{len(list_data)}): {item['title'][:20]}...")
            detail_data = fetch_detail_text(session, item["url"], delay=detail_delay)
            item.update(detail_data)
            all_data.append(item)

        time.sleep(list_delay)

    return all_data


# ==========================================
# 4. 데이터 분석 및 시각화 로직
# ==========================================
def extract_words(text, stopwords_set):
    """불용어를 필터링하고 2글자 이상의 의미 있는 명사/단어를 추출합니다."""
    text = re.sub(r"[^가-힣a-zA-Z0-9\s]", " ", text)
    words = text.split()
    return [w.lower() for w in words if len(w) >= 2 and w not in stopwords_set]


def visualize_data(df, font_path=FONT_PATH):
    """수집된 데이터를 바탕으로 워드클라우드 및 상위 빈도 그래프를 출력합니다."""
    if df.empty:
        print("[오류] 분석할 데이터프레임이 비어 있습니다.")
        return

    text_columns = ["title", "question_text", "answers_text", "detail_text"]
    available_columns = [col for col in text_columns if col in df.columns]
    
    df["combined_text"] = df[available_columns].fillna("").agg(" ".join, axis=1)
    all_text = " ".join(df["combined_text"].astype(str))

    words = extract_words(all_text, STOPWORDS)
    word_counts = Counter(words)
    top10_df = pd.DataFrame(word_counts.most_common(10), columns=["word", "count"])

    # 시각화 1: 워드클라우드
    wordcloud = WordCloud(
        font_path=font_path if os.path.exists(font_path) else None,
        width=1200, height=800,
        background_color="white"
    ).generate_from_frequencies(word_counts)

    plt.figure(figsize=(12, 6))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("네이버 지식인 텍스트 분석 (Word Cloud)", fontsize=16, pad=15)
    plt.show()

    # 시각화 2: 상위 10개 단어 막대그래프
    plt.figure(figsize=(12, 5))
    plt.bar(top10_df["word"], top10_df["count"], color="skyblue", edgecolor="gray")
    plt.title("가장 많이 등장한 상위 10개 단어", fontsize=14, pad=15)
    plt.xlabel("단어", fontsize=12)
    plt.ylabel("빈도수", fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()


# ==========================================
# 5. 프로그램 실행 진입점 (Main)
# ==========================================
if __name__ == "__main__":
    # 환경 설정 실행
    init_korean_font()

    # 데이터 수집 (크롤링 파이프라인 실행)
    raw_data = crawl_kin_pipeline(START_URL, total_pages=10, list_delay=1.2, detail_delay=0.8)
    data_df = pd.DataFrame(raw_data)

    if not data_df.empty:
        # 중복 데이터 제거 및 CSV 저장
        data_df = data_df.drop_duplicates(subset=["doc_id"])
        data_df.to_csv(SAVE_FILE, index=False, encoding="utf-8-sig")

        print("\n" + "=" * 40)
        print(f"[성공] 데이터 수집 및 저장이 완료되었습니다.")
        print(f" - 총 유니크 건수: {len(data_df)} 건")
        print(f" - 저장 위치: {os.path.abspath(SAVE_FILE)}")
        print("=" * 40 + "\n")

        # 데이터 시각화 실행
        visualize_data(data_df)
    else:
        print("[실패] 수집된 데이터가 존재하지 않습니다. 구조 변경 또는 웹 사이트 차단 여부를 확인하세요.")
```
